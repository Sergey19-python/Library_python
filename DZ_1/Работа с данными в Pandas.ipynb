```python
import pandas as pd, numpy as np
```

```python
authors = pd.DataFrame({ 'author_id':[1, 2, 3], 'author_name':["Тургенев", "Чехов", "Островский"]}, 
                      columns = ['author_id', 'author_name'])
```

```python
authors
```

```python
book = pd.DataFrame({"author_id":[1, 1, 1, 2, 2, 3, 3], 
                     "book_title":['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                    "price":[500, 400, 300, 350, 450, 600, 200]},
                   columns = ['author_id', 'book_title', 'price'])
```

```python
book
```

```python
authors_price = pd.merge(authors, book, on = 'author_id', how = "outer")
```

```python
authors_price
```

```python
# Берём первые пять строк  с max стоимостью
top5 = authors_price.nlargest(5, "price")
top5 = top5.reset_index(drop = True)
top5
```

```python
# В новом датафрейме создаём столбцы и высчитываем значения 
authors_stat = authors_price.groupby("author_name").agg({'price':['min', 'max', 'mean']}).reset_index()
authors_stat.columns = ['author_name','min_price', 'max_price', 'mean_price']
```

```python
authors_stat
```

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
book_info = pd.pivot_table(authors_price, 
                           index = ["author_name"], 
                           columns = ["cover"], 
                           values = ["price"],
                          aggfunc = np.sum,
                          fill_value = 0)
```

```python
book_info
```
